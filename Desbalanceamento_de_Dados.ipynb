## Importamos nosso DataSet

Observando nosso DataFrame, podemos chegar à conclusão de que o nosso conjunto de dados é composto por diversas informações sobre os nossos clientes, como:
* idade,
* trabalho (tipo de emprego), 
* estado civil, 
* nível de escolaridade, 
* default: tem crédito em default? (categórico: 'não','sim','desconhecido')
* balance: saldo
* housing (habitação, moradia)
* loan (empréstimo)
* contato: tipo de comunicação de contato (categórico: 'celular','telefone')
* day_of_week: último dia de contato da semana (categórico: 'mon','tue','wed','thu','fri')
* mês: último mês de contato do ano (categórico: 'jan', 'fev', ' mar', ..., 'nov', 'dec')
* duração: duração do último contato, em segundos (numérico).
*  campanha: número de contatos realizados durante esta campanha e para este cliente (numérico, inclui último contato)
* pdays: número de dias que se passaram após o último contato do cliente de uma campanha anterior contactado anteriormente)
* previous (anterior) número de contactos realizados antes desta campanha e para este cliente (numérico)
* poutcome (resultado): resultado da campanha de marketing anterior (categórica: 'fracasso','inexistente','sucesso')

* Variável de saída (desejada target):  "y" - o cliente subscreveu um depósito a prazo? (binário: 'sim', 'não')
